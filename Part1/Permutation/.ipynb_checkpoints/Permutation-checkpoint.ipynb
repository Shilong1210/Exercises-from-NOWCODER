```python
# -*- coding:utf-8 -*-
class Solution:
    def Permutation(self, ss):
        # write code here
        if(len(ss)==0):
            return []
        perm_list = []
        self.perm(ss,perm_list,'')
        perm_list = list(set(perm_list))
        return sorted(perm_list)
        
    def perm(self,ss,perm_list,path):
        if ss=='':
            perm_list.append(path)
        else:
            for i in range(len(ss)):
                self.perm(ss[:i]+ss[i+1:],perm_list,path+ss[i])
```
